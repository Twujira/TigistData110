```python
#Import necessary packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the dataset
url = "https://raw.githubusercontent.com/Twujira/TigistData110/main/Dataset/japan_quakes_01-18.csv"
Japanquake = pd.read_csv(url)
print(Japanquake.head())
```

```python
# Import necessary libraries
import pandas as pd

# Assuming JapanQuakes DataFrame is already loaded

# Convert the 'time' column to datetime format
Japanquake['time'] = pd.to_datetime(Japanquake['time'])

# Extract the year from the 'time' column and create a new column 'year'
Japanquake['year'] = Japanquake['time'].dt.year

# Filter the DataFrame to include only rows from the year 2011
filtered_JapanQuakes = Japanquake[Japanquake['year'] >= 2011]

# Display the first few rows of the filtered DataFrame
print(filtered_JapanQuakes.head())
print(filtered_JapanQuakes.tail())
```

```python
# Create a histogram for the 'mag' column
plt.figure(figsize=(6, 4))
sns.histplot(data=filtered_JapanQuakes, x='mag' ,bins=10, color='orange', alpha=1)

# Add labels and title
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.title('Histogram of Magnitude')

# Display the plot
plt.show()
```

```python
# Boxplot of magnitude by corresponding type
plt.figure(figsize=(6, 4))
sns.boxplot(x='magType', y='mag', data=filtered_JapanQuakes)

# Add labels and title
plt.xlabel('magType')
plt.ylabel('Magnitude')
plt.title('Box Plot of Magnitude by magType')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a 2D KDE plot with year on the x-axis and magnitude on the y-axis
plt.figure(figsize=(12, 8))

# Use a 2D KDE plot
sns.kdeplot(
    data=filtered_JapanQuakes,
    x='year',
    y='mag',  # Set magnitude on the y-axis
    hue='magType',  # Optional: Color by magType
    fill=True,  # Fill the plot for visual clarity
    cmap='viridis',  # Optional: Color map for the plot
    common_norm=False,  # Plot each group independently
)

# Set plot title and labels
plt.title('2D KDE Plot of Earthquake Magnitude over the Years')
plt.xlabel('Year')
plt.ylabel('Magnitude')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have filtered_JapanQuakes DataFrame with columns:
# 'latitude', 'longitude', 'magType', 'year', 'mag'

# Group the data by year and magnitude, then calculate the frequency of each combination
Quake_Group = filtered_JapanQuakes.groupby(['year', 'mag']).size().reset_index(name='frequency')

# Create a scatter plot with year on the x-axis and frequency of magnitude on the y-axis
plt.figure(figsize=(6, 4))
sns.scatterplot(
    data=Quake_Group,
    x='mag',
    y='frequency',
    hue='year',  # Optional: Color points by magnitude for better differentiation
    palette='viridis',  # Choose a color palette
    alpha=0.7,  # Set the transparency level of the points
)

# Set plot title and labels
plt.title('Scatter Plot of Frequency of Earthquake Magnitude by Year')
plt.xlabel('Magnitude')
plt.ylabel('Frequency of Magnitude')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have filtered_JapanQuakes DataFrame with columns:
# 'latitude', 'longitude', 'magType', 'year', 'mag'

# Group the data by year and magnitude, then calculate the frequency of each combination
Quake_Group = filtered_JapanQuakes.groupby(['year', 'mag']).size().reset_index(name='frequency')

# Pivot the data to create a grid for the heatmap
Quake_Pivot = Quake_Group.pivot(index='year', columns='mag', values='frequency').fillna(0)

# Create a heatmap to visualize the frequency of magnitudes by year
plt.figure(figsize=(10, 6))
sns.heatmap(
    data=Quake_Pivot,
    cmap='viridis',  # Choose a color palette
    annot=True,  # Show the frequency values in the heatmap
    fmt='g',  # Format the annotations as integers
)

# Set plot title and labels
plt.title('Heatmap of Frequency of Earthquake Magnitude by Year')
plt.xlabel('Magnitude')
plt.ylabel('Year')

# Display the heatmap
plt.show()
```

```python
# Set the style to white grid
sns.set(style="whitegrid")

# Set the figure size
plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered_JapanQuakes, x='mag', y='depth', hue='year', palette='Set1')

# Add a regression line
sns.regplot(data=filtered_JapanQuakes, x='mag', y='depth', scatter=False, color='blue')
plt.title('mag vs. depth in with Regression Line')
plt.grid(True)
plt.show()
```

Magnitude and depth have very weak correlation. as we can see in the correlation heat map below, the correlation of magnitude and depth is 0.07.

```python
# Select only the specified columns from the DataFrame
columns_new = ['latitude', 'longitude', 'depth', 'mag','gap','rms']

# Filter the DataFrame to include only the specified columns
filtered_JapanQuakes = filtered_JapanQuakes[columns_new]

# Calculate the correlation matrix for the selected numeric features
corr_matrix = filtered_JapanQuakes.corr()

# Set up the figure size
plt.figure(figsize=(6, 4))

# Visualize the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Add a title
plt.title('Correlation Matrix of Selected Features')

# Show the plot
plt.show()
```
